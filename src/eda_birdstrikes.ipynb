### __Clean Dataset__  

---

#### ___Create Date Columns___

```python
birdstrikes_clean = birdstrikes
birdstrikes_clean['Flight_Date'], _ = birdstrikes['Flight_Date'].str.split(' ',1).str
birdstrikes_clean['date'] = pd.to_datetime(birdstrikes_clean['Flight_Date'], format = '%m/%d/%y')
birdstrikes_clean = birdstrikes_clean.set_index('date')
birdstrikes_clean['year'] = birdstrikes_clean.index.year
birdstrikes_clean['month'] = birdstrikes_clean.index.month
birdstrikes_clean['day'] = birdstrikes_clean.index.day
birdstrikes_clean = birdstrikes_clean.reset_index()
```

#### ___Rename Columns___

```python
birdstrikes_clean = birdstrikes_clean.rename(columns = {'Effect__Amount_of_damage' : 'damage_level', 
                                                        'When__Time_of_day' : 'time_of_day',
                                                        'When__Phase_of_flight' : 'flight_phase',
                                                        'Wildlife__Size' : 'bird_size', 
                                                        'Origin_State' : 'state', 
                                                        'Airport__Name': 'airport',
                                                        'Aircraft__Airline_Operator': 'operator',
                                                        'Speed_IAS_in_knots' : 'speed',
                                                        'Aircraft__Make_Model' : 'aircraft_model', 
                                                        'Wildlife__Species' : 'species'})
```

#### ___Create Column for Specific `Damage Level` Ordering___

```python
birdstrikes_clean['damage_level_sort'] = 10
for i in range(birdstrikes_clean.shape[0]):
    if birdstrikes_clean.loc[i, 'damage_level'] == 'None':
        birdstrikes_clean.loc[i, 'damage_level_sort'] = 1
    elif birdstrikes_clean.loc[i, 'damage_level'] == "Minor":
        birdstrikes_clean.loc[i, 'damage_level_sort'] = 2
    elif birdstrikes_clean.loc[i, 'damage_level'] == "Medium":
        birdstrikes_clean.loc[i, 'damage_level_sort'] = 3
    elif birdstrikes_clean.loc[i, 'damage_level'] == "Substantial":
        birdstrikes_clean.loc[i, 'damage_level_sort'] = 4
    else:
        birdstrikes_clean.loc[i, 'damage_level_sort'] = 5
```

#### ___Select Columns___

```python
birdstrikes_clean = birdstrikes_clean[['date', 'year', 'month', 'day', 'time_of_day', 
                                       'damage_level', 'damage_level_sort', 'flight_phase',
                                       'state', 'airport', 'operator', 'speed', 'aircraft_model', 
                                       'bird_size', 'species']]
```

#### ___Filter Rows w/ Limited Observations___

```python
birdstrikes_clean = birdstrikes_clean.query('damage_level != "C" & damage_level != "B" & flight_phase != "Parked" & flight_phase != "Taxi"')
```

### __Export Cleaned Data Set__

```python
birdstrikes_clean.to_csv('../data/birdstrikes_clean.csv')
```
